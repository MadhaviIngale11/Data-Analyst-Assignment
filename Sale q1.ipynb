```python
# Create a new row of data (must match column structure)
new_row = {
    'Row ID': 9999,
    'Order ID': 'CA-2025-999999',
    'Order Date': '2025-01-10',
    'Ship Date': '2025-01-12',
    'Ship Mode': 'Second Class',
    'Customer ID': 'AB-99999',
    'Customer Name': 'Test Customer',
    'Segment': 'Consumer',
    'Country': 'United States',
    'City': 'Test City',
    'State': 'Test State',
    'Postal Code': 12345,
    'Region': 'West',
    'Product ID': 'OFF-TEST-999',
    'Category': 'Office Supplies',
    'Sub-Category': 'Binders',
    'Product Name': 'Test Product',
    'Sales': 250.00,
    'Quantity': 5,
    'Discount': 0.1,
    'Profit': 50.00
}

# Append to DataFrame
df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

print("After CREATE:")
display(df.tail(5))
```

```python
# Top 10 highest sales orders in California
california_top_sales = df[df['State'] == 'California'].sort_values(by='Sales', ascending=False).head(10)
display(california_top_sales)
```

```python
# Total sales by category
category_summary = df.groupby('Category')['Sales'].sum().reset_index()
display(category_summary)
```

```python
#Update Discount for a specific Order ID
df.loc[df['Order ID'] == 'CA-2025-999999', 'Discount'] = 0.2

print("After UPDATE (Discount changed):")
display(df[df['Order ID'] == 'CA-2025-999999'])
```

```python
#Delete the test row we added
df = df[df['Order ID'] != 'CA-2025-999999']

print("After DELETE:")
display(df.tail())
```
